## legends & visual styling

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,11,10)
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x);
axes.plot(x,x**2);
#  here which color represent which axes in not clearly defined.
```

```python
# adding legend to define which color represent which axes
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x,label="X vs X");
axes.plot(x,x**2,label="X vs X^2");
axes.legend(loc='lower left')
```

```python
#  setting manually
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x,label="X vs X");
axes.plot(x,x**2,label="X vs X^2");
axes.legend(loc=(1.1,0.5)) # reprsent first 2 parameters of add_axes()
```

```python
#  also can add -ve
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x,label="X vs X");
axes.plot(x,x**2,label="X vs X^2");
axes.legend(loc=(-0.5,0.5))
```

## Visual styling

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x,color='green',label="X vs X") # RDB Hex code 
axes.plot(x,x**2,color='#FF8293',label="X vs X^2",linewidth=4,ls="-.")   # ls="--","-.",":","-"
axes.plot(x,x**3,color='#C6FF4A',label="X vs X^3",lw=5,ls=":")   # lw -> line width  ls -> line style
axes.legend()
```

```python
# manual line spacing
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
lines=axes.plot(x,x**3,color='#C6FF4A',label="X vs X^3",lw=5)
lines[0].set_dashes([4,2,8,4,16,8]) # solid points,space point ,solid points, space points,,,,,,,,
```

```python
type(lines)
```

```python
#  styling marker  -> marker mark the points of x,y
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**3,color='#C6FF4A',label="X vs X^3",lw=2,ls="--",marker='v') #"o"-> circle , "v"-> triangle,,,,,,,
axes.plot(x,x**2,color='#FF8293',label="X vs X^2",linewidth=1,ls="-.",marker="+",markersize=10)
axes.legend()
```

```python
#  markers actually contains edges and center
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2,color='#FF8293',label="X vs X^2",linewidth=1,ls="-.",marker="o",markersize=10,
          markerfacecolor='red',markeredgewidth=3,markeredgecolor='purple')
```
