```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,10)
y=2*x
```

```python
# figure object
fig=plt.figure()
axes=fig.add_axes([0.5,0.5,0.3,0.3])
axes.plot(x,y)
```

## adding multiple axes

```python
a=np.linspace(0,10,11)
```

```python
a
```

```python
b=a**4
```

```python
b
```

```python
fig=plt.figure()

# large axes
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b);
axes1.set_xlabel('X axis')
axes1.set_ylabel('Y axis')
axes1.set_title('Power of 4')

# small axes
axes2=fig.add_axes([0.15,0.5,0.3,0.4])

axes2.plot(a,b);
axes2.set_xlim(1,3)
axes2.set_ylim(0,100)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')
```

```python
type(fig)
```

```python
#  dpi -> dotes per inch (increasing resolution)   
fig=plt.figure(figsize=(2,2),dpi=200)
axes=fig.add_axes([0,0,1,1])
axes.plot(a,b);

# setting bounding box inches to tight to show x-y axes tick numbers in image
fig.savefig('new_figure.png',bbox_inches='tight')
```
