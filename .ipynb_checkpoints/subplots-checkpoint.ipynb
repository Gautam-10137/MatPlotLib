## builtin method to have subplots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.linspace(0,10,11)
b=a**4
```

```python
x=np.arange(0,10)
y=2*x
```

```python
fig,axes=plt.subplots()
axes.plot(x,y)
```

```python
type(axes)  # object
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
# here axes is a numpy array containing 2 objects

axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
axes     # numpy array
```

```python
# 2 dimensional array
fig,axes=plt.subplots(2,2)
# axes[0].plot(x,y)           have to take as 2d

axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
```

```python
axes.shape
```

```python
# looping through each plot
fig,axes=plt.subplots(3,1)
for ax in axes:
    ax.plot(x,y)
# to adjust spacing
plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(4,6),dpi=200)
axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][0].set_xlabel("X axis")
axes[1][1].plot(a,b)

# overall sup title with custom font size
fig.suptitle("Figure Level",fontsize=18)

# manual spacing
fig.subplots_adjust(wspace=1,hspace=1)

# applicable to each figure
fig.set_figwidth(10)

# saving
fig.savefig("new_subplots.png",bbox_inches='tight')
```
